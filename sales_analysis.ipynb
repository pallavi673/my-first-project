```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Project folder
project_folder = r"C:\Users\rawat\OneDrive\Desktop\sales project"

# CSV file
csv_filename = "train.csv"
csv_path = os.path.join(project_folder, csv_filename)

# Load CSV
df = pd.read_csv(csv_path)
print("First 5 rows of the dataset:")
display(df.head())

# Dataset info
print("\nDataset info:")
df.info()

# Summary statistics
print("\nSummary statistics:")
display(df.describe())

# Missing values
print("\nMissing values:")
print(df.isnull().sum())

# Drop missing values
df = df.dropna()

# Total Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=sales_by_region)
plt.title('Total Sales by Region')
plt.show()

# Top 5 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10,5))
top_products.plot(kind='bar', title='Top 5 Products by Sales')
plt.ylabel("Sales")
plt.show()

# Sales Trend Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
sales_over_time = df.groupby('Order Date')['Sales'].sum().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(x='Order Date', y='Sales', data=sales_over_time)
plt.title('Sales Trend Over Time')
plt.show()

# Category-wise Sales
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=sales_by_region)
plt.title('Total Sales by Region')
plt.savefig(r"C:\Users\rawat\OneDrive\Desktop\sales project\sales_by_region.png")  # Saves as PNG
plt.show()
sales_by_region.to_csv(os.path.join(project_folder, "sales_by_region.csv"), index=False)
category_sales.to_csv(os.path.join(project_folder, "category_sales.csv"), index=False)
```
